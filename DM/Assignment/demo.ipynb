# Step-1 Import CSV 

```python
import pandas as pd
data = pd.read_csv("./heart.csv")
data
```

# Step - 2 Print all the details

```python
data.info()
```

# Step-3 how many patients have heart disease versus those who do not?

```python
data['target'].value_counts()
```

# Step-4 What is the average age of patients with heart disease?

```python
average_age_hd = data[data['target'] == 1]['age'].mean()
print(average_age_hd)
```

# Step-5 How many patients have high blood pressure?

```python
high_bp_count = data[data['trestbps'] >= 140].shape[0]
print(high_bp_count)
```

# Step-6 What is the average cholesterol level of patients with heart disease?

```python
average_chol_hd = data[data['target'] == 1]['chol'].mean()
print(average_chol_hd)
```

# Step-7 How many males have high blood pressure?

```python
male_high_bp_count = data[(data['trestbps'] >= 140) & (data['sex'] == 1)].shape[0]
print(male_high_bp_count)
```

# Step-8 Visualize the distribution of heart disease (target) in the dataset using a bar chart.

```python
import matplotlib.pyplot as plt

data['target'].value_counts().plot(kind='bar')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Number of Patients')
plt.title('Distribution of Heart Disease in Dataset')
plt.show()
```

# Step-9 Visualize the distribution of patients with heart disease by age using a histogram.

```python

plt.figure(figsize=(10, 6))
plt.hist(data[data['target'] == 1]['age'], bins=30, alpha=0.7, color='red', label='Heart Disease')
plt.hist(data[data['target'] == 0]['age'], bins=30, alpha=0.7, color='blue', label='No Heart Disease')
plt.title('Distribution of Patients with Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.legend()
plt.show()
```

# Step-16 Find the max age of patients with heart disease

```python
max_age_hd = data[data['target'] == 1]['age'].max()
print(f"Maximum age of patients with heart disease: {max_age_hd}")
```

# Step-17 Find the min age of patients with heart disease

# Step-18 Find the average maximum heart rate (thalach) for patients with heart disease

```python
average_thalach_hd = data[data['target'] == 1]['thalach'].mean()
print(f"Average maximum heart rate for patients with heart disease: {average_thalach_hd}")
```

# Step-19. How many patients have a cholesterol level above 300?

```python
high_chol_count = data[data['chol'] > 300].shape[0]
print(f"Number of patients with cholesterol > 300: {high_chol_count}")
```

# Step-20. What is the average resting blood pressure (trestbps) for patients without heart disease?


```python
avg_trestbps_no_hd = data[data['target'] == 0]['trestbps'].mean()
print(f"Average resting blood pressure (no heart disease): {avg_trestbps_no_hd}")
```

# Step-21 What is the average maximum heart rate (thalach) for patients without heart disease?


```python
avg_thalach_no_hd = data[data['target'] == 0]['thalach'].mean()
print(f"Average maximum heart rate (no heart disease): {avg_thalach_no_hd}")
```

# Step-22. What is the average cholesterol level for males vs females?

```python
avg_chol_male = data[data['sex'] == 1]['chol'].mean()
avg_chol_female = data[data['sex'] == 0]['chol'].mean()
print(f" male: {avg_chol_male}")
print(f" female: {avg_chol_female}")
```

# Step-23. What is the distribution of chest pain types (cp) among patients?


```python
print(data['cp'].value_counts())
```

# Step-24: Create a boxplot to compare cholesterol levels (chol) between males and females.


```python
plt.figure(figsize=(8, 6))
data.boxplot(column='chol', by='sex', grid=False)
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Cholesterol Level')
plt.title('Cholesterol Levels by Sex')
plt.suptitle('')
plt.show()
```

# Step-25: What is the distribution of cholesterol levels among patients with and without heart disease?

```python
plt.figure(figsize=(10, 6))
plt.hist(data[data['target'] == 1]['chol'], bins=30, alpha=0.7, color='red', label='Heart Disease')
plt.hist(data[data['target'] == 0]['chol'], bins=30, alpha=0.7, color='blue', label='No Heart Disease')
plt.title('Cholesterol Level Distribution by Heart Disease Status')
plt.xlabel('Cholesterol Level')
plt.ylabel('Number of Patients')
plt.legend()
plt.show()
```
